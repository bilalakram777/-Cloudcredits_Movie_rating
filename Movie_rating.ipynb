```python



import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse



from google.colab import files
uploaded = files.upload()


ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')


reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)


trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

model = SVD()
model.fit(trainset)


predictions = model.test(testset)


print("Model Evaluation:")
rmse(predictions)


user_id = 1


all_movie_ids = set(movies['movieId'])


rated_movies = set(ratings[ratings['userId'] == user_id]['movieId'])


unrated_movies = all_movie_ids - rated_movies


unrated_predictions = [
    (movie_id, model.predict(user_id, movie_id).est)
    for movie_id in unrated_movies
]


recommended = sorted(unrated_predictions, key=lambda x: x[1], reverse=True)[:10]


print("\nTop 10 movie recommendations for user ID", user_id)
for movie_id, predicted_rating in recommended:
    movie_title = movies[movies['movieId'] == movie_id]['title'].values[0]
    print(f"{movie_title} (Predicted Rating: {predicted_rating:.2f})")
```
